# Effects of Loan Characteristics on Their Investors 


### Investigation Overview: 
In this investigation, I wanted to look at the characteristics of loans that could be used to predict their I expect that the total loan amount will have a positive effect on the Investors. I also think that the EmploymentStatus, LoanStatus, BorrowerAPR, IncomeRange, CreditGrade will also have effects on the Investors.

### Dataset Overview
The dataset contains 113,937 loans with 81 features including (EmploymentStatus, LoanStatus, BorrowerAPR, IncomeRange, CreditGrade, Investors and many others features)


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
# load in the dataset into a pandas dataframe, print statistics
df = pd.read_csv('prosperLoanData.csv')
```

```python
features = ['EmploymentStatus', 'LoanStatus', 'BorrowerAPR', 'IncomeRange', 'CreditGrade', 'Investors']
df_subset = df[features]
```

### Distribution of Borrower APR

##### The distribution of Investors looks a right-skewed : A large peak centered at 0 - 200.  Only very few loans have Investors in 600.

```python
bins = np.arange( 0, df_subset['Investors'].max()+50, 50)

plt.figure(figsize=[8, 5])
plt.hist(data = df_subset, x = 'Investors', bins = bins)
plt.xlabel('Investors')
plt.ylabel(' Count')
plt.title('Count of Investors')
plt.show()
```

### Distribution of Borrower APR
#### The BorrowerAPR has A random distribution looks,  a large peak centered at 0.1 - 0.2.

```python
binsize = 0.05
bins = np.arange(0, df_subset['BorrowerAPR'].max()+binsize, 0.01)

plt.figure(figsize=[8, 5])
plt.hist(data = df_subset, x = 'BorrowerAPR', bins = bins)
plt.xlabel(' BorrowerAPR')
plt.ylabel(' Count')
plt.title('Count of BorrowerAPR')
plt.show()
```

### Investors vs. BorrowerAPR
##### This plot shows that at different size for investors, the APR has a large range, but the APR range decreases as investors increase. In general, the borrower's APR is negatively correlated with the investors.

```python
plt.figure(figsize = [8, 5])
ax = sb.regplot(data = df_subset, x = 'Investors', y = 'BorrowerAPR', scatter_kws={'alpha':1/10});
plt.title('Rlation between the BorrowerAPR and Investors')
```

### Borrower APR by CreditGrade and EmploymentStatus
##### Here we see that the borrower's APR of class HR (CreditGrade) for retired employees is increasing, while the borrower's APR of class AA and class A (CreditGrade) is decreasing.

```python
fig = plt.figure(figsize = [10,10])

ax = sb.pointplot(data = df_subset, x = 'CreditGrade', y = 'BorrowerAPR', hue = 'EmploymentStatus',
           palette = 'Greens', linestyles = '', dodge = 0.4)
plt.title('Borrower APR across EmploymentStatus and CreditGrade')
plt.ylabel('Borrower APR')

ax.set_yticklabels([],minor = True);
```
